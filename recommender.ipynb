```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/processed_sales_data.csv')
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
customers = data.customer_id.unique()
```

```python
df = {}
for customer in customers:
    rec = []
    for i in range(len(data)):
        if customer == data['customer_id'][i]:
            rec.append(data['item_id'][i])
        df[customer] = rec
```

```python
df
```

```python
new_data = pd.DataFrame(df.items(), columns=['customer', 'item'])
```

```python
new_data.shape
```

```python
new_data
```

```python
new_data['customer'] = new_data['customer'].apply(lambda val : val.replace('customer', ''))
```

```python
new_data
```

```python
new_data['item'] = new_data['item'].apply(lambda val : [v.replace('Item_', '') for v in val])
```

```python
new_data
```

```python
new_df = pd.melt(new_data.set_index('customer')['item'].apply(pd.Series).reset_index(),
        id_vars=['customer'],
        value_name='item'
        ).dropna().drop(['variable'], axis=1) \
    .groupby(['customer', 'item']) \
    .agg({'item' : 'count'}) \
    .rename(columns={'item' : 'purchase_count'}) \
    .reset_index() \
    .rename({'item' : 'productID'}, axis=1)
```

```python
new_df
```

```python
new_df['customer'] = new_df['customer'].apply(lambda x: int(x))
new_df['productID'] = new_df['productID'].apply(lambda x: int(x))
new_df['purchase_count'] = new_df['purchase_count'].apply(lambda x: int(x))
```

```python
rating_matrix = new_df.pivot_table(values='purchase_count', index='customer', columns='productID', fill_value=0)
```

```python
rating_matrix
```

```python
X = rating_matrix.T.copy()
```

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
```

```python
decomposed_matrix.shape
```

```python
import numpy as np
corr = np.corrcoef(decomposed_matrix)
```

```python
corr
```

```python
prod = 1
customer_ids = list(X.index)
customer_id = customer_ids.index(prod)
corr_prod_id = corr[customer_id]
recommend = list(X.index[corr_prod_id > 0.5])
recommend.remove(prod)

recommend
```

```python
def recommend_items(id):
    prod = id
    customer_ids = list(X.index)
    customer_id = customer_ids.index(prod)
    corr_prod_id = corr[customer_id]
    recommend = list(X.index[corr_prod_id > 0.5])
    recommend.remove(prod)
    return recommend
```
