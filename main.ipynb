```python
import pandas as pd
```

```python
dataset = pd.read_csv('dataset/Sales_Data.csv')
```

```python
dataset
```

```python
df_dict = {}
df_dict['date'] = []
df_dict['customer_id'] = []
df_dict['item_id'] = []
df_dict['quantity'] = []
df_dict['price_per_unit'] = []
df_dict['amount'] = []
df_dict['Vrh_No'] = []

current_date = None
current_customer = None
current_Vrh_No = 0
for i, row in enumerate(dataset.iterrows()):
    row = row[1]
    if row['Particulars'] == 'By' or not row['Particulars'] == 'To':
        if isinstance(row['Date'], str):
            current_date = row['Date']
            current_customer = row['Name']
            current_Vrh_No += 1
        else:
            if isinstance(row['Name'], float): break
            df_dict['date'].append(current_date)
            df_dict['customer_id'].append(current_customer)
            df_dict['Vrh_No'].append(current_Vrh_No)
            df_dict['item_id'].append(row['Name'])
            df_dict['quantity'].append(row['Vch Type'])
            df_dict['price_per_unit'].append(row['Vch No.'])
            df_dict['amount'].append(row['Debit'])
    
```

```python
size = len(df_dict['date'])
print(size)
for key in df_dict.keys():
    if len(df_dict[key]) != size:
        print(key, len(df_dict[key]))
        assert False
```

```python
df = pd.DataFrame(df_dict)
```
